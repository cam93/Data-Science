## The goal of this project is to track the exchange rates between the US dollar and the Euro in recent years.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

 - We rename the [US dollar ] and Period\Unit: columns to something easier to type â€” US_dollar and Time.
 - We change the Time column to a datetime data type.
 - We sort the values by Time in ascending order.
 - We reset the index (and drop the initial index).

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar']= euro_to_dollar['US_dollar'].astype(float)
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year>=2006)
                                         & (euro_to_dollar['Time'].dt.year <= 2009)]
fc_0809 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007)
                               & (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

fig,ax =plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],financial_crisis['rolling_mean'],linewidth=1, color='#A6D785')

ax.plot(fc_0809['Time'], fc_0809['rolling_mean'], linewidth=3, color='#e23d28')

ax.axvspan(xmin=733112.0, xmax=733302.0,ymin=0.09,alpha=0.3,color='grey')


ax.grid(alpha=0.5)
ax.set_xticklabels([])
ax.set_yticklabels([])

x=732272.0
for year in ['2006','2007','2008','2009','2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365
y=1.193
for rate in ['1.2','1.3','1.4','1.5']:
    ax.text(732172.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
ax.text(732172.0, 1.67, "Euro-USD rate peaked at 1.59 during 2007-08 finacial crisis", weight='bold')
ax.text(732172.0, 1.63, "Euro-USD exchange rates between 2006 and 2010", size=12)


plt.show()
```
